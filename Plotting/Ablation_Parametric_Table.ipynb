```python
import wandb
```

```python
api = wandb.Api()
```

```python
sweeps = [api.sweep(f"Jorgvt/PerceptNet_v15/{sweep_id}") for sweep_id in sweep_ids]
runs = [sweep.runs for sweep in sweeps]
runs = [run for runs_ in runs for run in runs_]
len(runs)
```

```python
cols = ["train_loss", "val_loss"]
config_keys = ["PARAM_CS", "PARAM_DN_CS", "PARAM_GABOR", "PARAM_DN_FINAL", "FINAL_B"]
```

```python
def get_name(serie):
    params = []
    if serie.USE_GAMMA: params.append("1")
    if serie.PARAM_CS: params.append("4")
    if serie.PARAM_DN_CS: params.append("5")
    if serie.PARAM_GABOR: params.append("6")
    if serie.PARAM_DN_FINAL: params.append("7")
    if serie.FINAL_B: params.append("8")
    return "[" + ",".join(params) + "]"
```

```python
run.sweepName
```

```python
from tqdm.auto import tqdm

histories = []
for run in tqdm(runs):
    history = run.history()
    if len(history) < 500: continue

    history = history[cols]
    history["id"] = run.id
    history["sweep_id"] = run.sweepName
    history[config_keys] = [run.config[key] for key in config_keys]
    history["Epoch"] = range(len(history))
    if not hasattr(history, "USE_GAMMA"): history["USE_GAMMA"] = True

    ## Extract best values during training
    history = history.filter(["train_loss", "val_loss", "id", "sweep_id", "USE_GAMMA", "PARAM_CS", "PARAM_DN_CS", "PARAM_GABOR", "PARAM_DN_FINAL", "FINAL_B"])\
                     .min()
    history["name"] = get_name(history)
    history["params"] = run.summary.total_parameters
    # break

    histories.append(history)
    # break
```

```python
import pandas as pd
histories = pd.DataFrame(histories)
histories
```

```python
histories["len"] = histories.name.apply(lambda x: len(x))
histories = histories.sort_values("len")
histories
```

```python
histories = histories.sort_values(by="val_loss", ascending=False)\
         .drop_duplicates(subset="name", keep="last")\
         .sort_values(by="len")
histories
```

```python
histories.to_csv("results_ablation_parametric_table.csv", index=False)
```
